                          What is WAF?

A WAF (Web Application Firewall) consists of a security solution that aims at the protection of web applications against online threats. Usually some of them are SQL Injection, file inclusion, cross-site scripting (XSS) and cross-site request forgery. It's integrated with Azure Application Gateway, Azure Front Door, and Azure Content Delivery Network (CDN) to provide centralized protection for web applications hosted on Azure.It works like a barrier between the net and the app: monitoring and filtering HTTP traffic. Azure WAF

Azure Web Application Firewall (WAF) consists of a cloud-based security service created by Microsoft Azure that helps protect web applications from common web threats and vulnerabilities. It's based-on Core Rule Set (CRS) pertaining to OWASP, the Open Web Application Security Project. Azure WAF features work within policies behind the application gateway. The Application Gateway operates as an ADC (Application Delivery Controller) providing services such as: transport layer security; cookie-based session affinity; multiple hosting; enforce security policies, inspect incoming traffic for malicious payloads, and mitigate attacks in real-time.

Capabilities  

In terms of capabilities Azure WAF can be understood through 3 categories: protection, monitoring and customization. It is capable of OWASP Protection enforcing compliance with the Open Web Application Security Project (OWASP) guidelines by blocking known attack vectors outlined by OWASP, such as injection attacks, security misconfigurations, and sensitive data exposure.several practices listed by Microsoft Azure would be:  

    Protect your web applications from web vulnerabilities and attacks without modification to back-end code.

    Protect multiple web applications at the same time. An instance of Application Gateway can host up to 40 websites that are protected by a web application firewall.

    Create custom WAF policies for different sites behind the same WAF.

    Protect your web applications from malicious bots with the IP Reputation ruleset.

    Protect your application against DDoS attacks.  

    QL injection protection.

    Cross-site scripting protection.

    Protection against other common web attacks, such as command injection, HTTP request smuggling, HTTP response splitting, and remote file inclusion.

    Protection against HTTP protocol violations.

    Protection against HTTP protocol anomalies, such as missing host user-agent and accept headers.

    Protection against crawlers and scanners.  

    Threat Intelligence: It leverages threat intelligence feeds to stay updated on the latest attack signatures and patterns, enabling it to effectively identify and block new and emerging threats.



Monitoring  

Azure Monitor is the service that in joint venture with Azure WAF provides monitoring capabilities. It collects metrics related to WAF deployment like request count, request size, HTTP status codes, and WAF rule matches. Producing insights into the performance of the service as well as the condition of WAF configuration. Logs are collected to encompass data as incoming HTTP requests, including the source IP address, destination URL, rule set name, action taken (allowed, blocked, or logged), etc.  

In addition, Azure monitor could manage alerts under predefined conditions or custom queries against WAF metrics and logs. For example, you can create alerts for unusually high request rates, many WAF rule matches, or specific types of attacks detected by the WAF. These alerts can notify you via email, SMS when certain thresholds are exceeded or when specific patterns are detected. An image of visualizing with Azure Monitor Logs is provided below. Some other monitoring uses might include:  

    Detection of common application misconfigurations (for example, Apache and IIS).

    Geo-filter traffic to allow or block certain countries/regions from gaining access to your applications.

    Inspect JSON and XML in the request body

Customization  

Azure WAF counts with customization capabilities so that WAF rules and rule groups can suit an application requirement and eliminate false positives. Also, it is possible to associate a WAF Policy for each site behind your WAF to allow for site-specific configuration. This means not only it is capable of deploying a policy with specificity but also simultaneously with different policies over the same traffic. There are two types of rules:  

    Custom rules that you create

    Managed rule sets that are a collection of Azure-managed preconfigured set of rules. These rules can be based on IP addresses, HTTP headers, request and response body content, or other parameters to provide granular control over traffic filtering. It integrates with Azure Monitor and Azure Security Center for centralized logging and monitoring.

If happens the case that both rules are present, then the customized rule will be processed before the managed ruleset. A rule consists of condition, priority and an action. Action types are ALLOW, BLOCK, and LOG. Priority is a unique integer that defines the order of rules to be processed. Smaller integer values are assigned to a higher order rule and high order rules are evaluated first. A Web application by application Gateway can have a WAF policy associated to it at the global level, per-site level and per-URI level. Being a cloud-based service, Azure WAF offers scalability and high performance to handle varying levels of web traffic and protect web applications against distributed denial-of-service (DDoS) attacks.The working process in brief steps would be:

- When a request is made to a protected web application, it first passes through the Azure WAF.
- Azure WAF inspects the request against its predefined rules and custom rules configured by the user.
- If the request is deemed malicious or violates any security policies, Azure WAF blocks it and returns an appropriate response (e.g., a block page or an HTTP error).
- Legitimate requests that pass the security checks are forwarded to the web application.
- Azure WAF continuously monitors traffic and adjusts its security policies to adapt to new threats and attack patterns, providing ongoing protection for web applications hosted on Azure.
